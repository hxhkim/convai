```python
# expected output format
transformed_data = {
    "participant_persona": {
        "participant_1": {
            # "id": "participant_1_id", # column for RDBS
            "name": "participant_1_name",
            "age": "participant_1_age",
            "gender": "participant_1_gender",
            "personality": bot_profile,
            "background": "participant_1_background"
        },
        "participant_2": {
            # "id": "participant_2_id", # column for RDBS
            "name": "participant_2_name",
            "age": "participant_2_age",
            "gender": "participant_2_gender",
            "personality": user_profile,
            "background": "participant_2_background"
        }
    },
    "messages": []
}

# transforme to the expected format
for i, message in enumerate(dialog):
    role = "participant_1" if message['sender'] == 'participant1' else "participant_2"
    # role = "user" is message['sender_class'] == 'Human' else "assistant" # role 값과 utterer 값을 분리할 경우 
    # utterer = "participant_1_name" if role == "participant_1" else "participant_2" # for RDBS
    
    transformed_message = {
        "role": role,
        "content": {
            # "utterance_id": str(i + 1),
            # "utterer": utterer, # for RDBS
            "emotion_scores" : "emotion_scores",
            "text": message['text'],
            "topic": "topic"
            },
        }
    
    transformed_data["messages"].append(transformed_message)

# save the transformed data
with open('/home/user1/conversation-data/dataset-01-convai/data/02_transformed_data/transfomed_data_intermediate.json', 'w', encoding='utf-8') as f:
    json.dump(transformed_data, f, indent=4, ensure_ascii=False)
```

### 최종 변환 코드

```python
import os
import json
from pprint import pprint

# read file
with open('/home/user1/conversation-data/dataset-01-convai/data/01_formatted_data/formatted_data_intermediate.json', 'r') as file:
    data = json.load(file)

for i in range(len(data)):
    dialog = data[i]['dialog']
    # print(f"Dialog {i + 1}")
    # pprint(dialog)

    bot_profile = " ".join(data[i].get('bot_profile', []))
    user_profile = " ".join(data[i].get('user_profile', []))

    # print(f"Dialog {i + 1}")
    # print(bot_profile)
    # print(user_profile)
    # print("----------------------------")


    transformed_data = {
    "participant_persona": {
        "participant_1": {
            # "id": "", # id column for RDBS
            "name": "",
            "age": "",
            "gender": "",
            "personality": "",
            "background": bot_profile
        },
        "participant_2": {
            # "id": "", # id column for RDBS
            "name": "",
            "age": "",
            "gender": "",
            "personality": "",
            "background": user_profile
        }
    },
    "messages": []
}
    
    # print(f"Dialog {i + 1}")    
    # pprint(transformed_data)

    for j, message in enumerate(dialog):
        role = "participant_1" if message['sender'] == 'participant1' else "participant_2"
        # role = "user" is message['sender_class'] == 'Human' else "assistant" # role 값과 utterer 값을 분리할 경우 
        # utterer = "participant_1_name" if role == "participant_1" else "participant_2" # for RDBS and groupchat
        
        transformed_message = {
            "role": role,
            "content": {
                # "utterance_id": str(j + 1),
                # "utterer": utterer, # for RDBS
                "emotion_scores" : {
                },
                "text": message['text'],
                },
            }
        
        transformed_data["messages"].append(transformed_message)


    output_dir = '/home/user1/conversation-data/dataset-01-convai/data/02_transformed_data/intermediate'
    output_file = os.path.join(output_dir, f'transfomed_data_intermediate_{i + 1}.json')
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(transformed_data, f, indent=4, ensure_ascii=False)
```
