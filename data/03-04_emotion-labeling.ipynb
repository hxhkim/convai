### filled data -> emotion labeling

```python
import openai
import json
import os
from typing import Dict
from dotenv import load_dotenv

load_dotenv()

MODEL = "gpt-4"
openai.api_key = os.getenv("OPENAI_API_KEY")


def analyze_emotion(data: Dict) -> Dict:
    for message in data['messages']:
        content = message.get('content', {})
        if not content.get('emotion_scores') and 'text' in content:
            text = content['text']
            prompt = f"""Analyze the following text and provide emotion scores for the following categories: Anger, Fear, Joy, Sadness, Surprise, Love, Boredom, Neutral. The output should be in JSON format with the emotion categories as keys and their respective scores as values, totaling exactly 100, with one decimal place precision. Ensure a balanced distribution that reflects the most likely emotional state based on the text content and the participant persona information.

            Expected output example:

            "emotion_scores": {{
                "Anger": 0,
                "Fear": 5,
                "Joy": 10,
                "Sadness": 5,
                "Surprise": 10,
                "Love": 5,
                "Boredom": 5,
                "Neutral": 60
            }},
            "text": "{text}""" 

            response = openai.ChatCompletion.create(
                model=MODEL,
                messages=[
                    {"role": "system", "content": "You are an emotion analysis assistant."},
                    {"role": "user", "content": prompt}
                ],
                temperature=0.8,
                response_format={"type": "json_object"}
            )
            
            response_content = response.choices[0].message.content
            print(response_content)

            # JSON 유효성 검사
            try:
                filled_data = json.loads(response_content)  
            except json.JSONDecodeError as e:
                print(f"JSONDecodeError: {e}")
                continue

            for message in data['messages']:
                content = message.get('content', {})
                if content.get('text') == filled_data.get('text'):
                    for key, value in filled_data.items():
                        content[key.lower()] = value
                    break           

    return data

with open('/home/user1/conversation-data/dataset-01-convai/data/03_filled_data/volunteers/filled_data_volunteers_1100.json', 'r') as file:
    data = json.load(file)

filled_data = analyze_emotion(data)

with open("/home/user1/conversation-data/dataset-01-convai/data/04_emotion_labeled_data/volunteers/emotion_labeled_data_volunteers_1100.json", "w") as file:
    json.dump(data, file, indent=4)
```

```python
import openai
import json
import os
import pprint
from dotenv import load_dotenv
from typing import Dict, List

load_dotenv()

MODEL = "gpt-4o-mini"
openai.api_key = os.getenv("OPENAI_API_KEY")

client = openai
client

def label_emotion(data: Dict) -> Dict:

    for message in data['messages']:
        content = message.get('content', {})
        if not content.get('emotion_scores') and 'text' in content:
            text = content['text']
            prompt = f"""Analyze the following text and provide emotion scores for the following categories: Anger, Fear, Joy, Sadness, Surprise, Love, Boredom, Neutral. The output should be in JSON format with the emotion categories as keys and their respective scores as values, totaling exactly 100. 

            # Guidelines:
            1. Consider the Participant Information and Conversation Log when interpreting the emotional content of the text.
            2. Subtle emotional cues should be reflected in the corresponding emotion scores, but don't overinterpret.
            3. Maintain a high Neutral score as the default. Only assign scores to other emotions when there is clear and strong evidence in the text.

            # Participant Information: 
            {data["participant_persona"]},

            # Conversation Log:
            {data["messages"]}
        
            # Expected output example:
            "emotion_scores": {{
                "Anger": ,
                "Fear": ,
                "Joy": ,
                "Sadness": 
                "Surprise": ,
                "Love": ,
                "Boredom": ,
                "Neutral": 
            }},
            "text": "{text}""" 

            response = client.ChatCompletion.create(
                model=MODEL,
                messages=[
                    {"role": "system", "content": "You are an AI assistant that helps to build conversation data set."},
                    {"role": "user", "content": prompt}
                ],
                temperature=0.8,
                response_format={"type": "json_object"}
            )

            response_content = response.choices[0].message.content

            # JSON 유효성 검사
            try:
                filled_data = json.loads(response_content)  
            except json.JSONDecodeError as e:
                print(f"JSONDecodeError: {e}")
                continue

            # 기존 데이터 업데이트
            if content.get('text') == filled_data.get('text'):
                for key, value in filled_data.items():
                    content[key.lower()] = value

    return data

with open('/home/user1/conversation-data/dataset-01-convai/data/03_filled_data/volunteers/filled_data_volunteers_30.json', 'r') as file:
    data = json.load(file)

filled_data = label_emotion(data)
print("filled_data:", filled_data)  # 실시간으로 받아오는 내역 출력

output_file_path = "/home/user1/conversation-data/dataset-01-convai/data/04_emotion_labeled_data/volunteers/emotion_labeled_data_volunteers_30_v4_updated.json"
with open(output_file_path, 'w', encoding='utf-8') as file:
    json.dump(data, file, ensure_ascii=False, indent=4)

    # Check if emotion_scores are empty and retry if necessary
    for message in data['messages']:
        content = message.get('content', {})
        if not content.get('emotion_scores') or not any(content['emotion_scores'].values()):
            text = content['text']
            prompt = f"""Analyze the following text and provide emotion scores for the following categories: Anger, Fear, Joy, Sadness, Surprise, Love, Boredom, Neutral. The output should be in JSON format with the emotion categories as keys and their respective scores as values, totaling exactly 100. 

            # Guidelines:
            1. Consider the Participant Information and Conversation Log when interpreting the emotional content of the text.
            2. Subtle emotional cues should be reflected in the corresponding emotion scores, but don't overinterpret.
            3. Maintain a high Neutral score as the default. Only assign scores to other emotions when there is clear and strong evidence in the text.

            # Participant Information: 
            {data["participant_persona"]},

            # Conversation Log:
            {data["messages"]}
        
            # Expected output example:
            "emotion_scores": {{
                "Anger": ,
                "Fear": ,
                "Joy": ,
                "Sadness": 
                "Surprise": ,
                "Love": ,
                "Boredom": ,
                "Neutral": 
            }},
            "text": "{text}""" 

            response = client.ChatCompletion.create(
                model=MODEL,
                messages=[
                    {"role": "system", "content": "You are an AI assistant that helps to build conversation data set."},
                    {"role": "user", "content": prompt}
                ],
                temperature=0.8,
                response_format={"type": "json_object"}
            )

            response_content = response.choices[0].message.content

            # JSON 유효성 검사
            try:
                filled_data = json.loads(response_content)  
            except json.JSONDecodeError as e:
                print(f"JSONDecodeError: {e}")
                continue

            # 기존 데이터 업데이트
            if content.get('text') == filled_data.get('text'):
                for key, value in filled_data.items():
                    content[key.lower()] = value

    # 업데이트된 데이터를 파일에 저장
    output_file_path = "/home/user1/conversation-data/dataset-01-convai/data/04_emotion_labeled_data/volunteers/emotion_labeled_data_volunteers_30_v5-2-rerequest_updated.json"
    with open(output_file_path, 'w', encoding='utf-8') as file:
        json.dump(data, file, ensure_ascii=False, indent=4)
```
